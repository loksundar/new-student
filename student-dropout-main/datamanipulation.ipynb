```python
import numpy as np
import pandas as pd
dat=pd.read_csv('student.csv')
dat.head()
```

```python
dat.columns
```

```python
dat.shape
```

```python
dat=dat.dropna()#removing na
```

```python
dat.shape
```

```python
19100 - 18762 #we removed these many values
```

```python
atr=[]

for x in dat['continue_drop']:
    if x=="continue":
        atr.append(0)
    else:
        atr.append(1)
    
atr
```

```python
dat['continue_drop']=atr
```

```python
dat.head()
```

```python
gen=[]
for x in dat['gender'] :
    if x=="M" :
        gen.append(1)
    else :
        gen.append(2)
gen    
```

```python
dat['gender'].value_counts()
```

```python
dat['gender']=gen
dat.head()
```

```python
dat['internet'].value_counts()
```

```python
inter=[]
for x in dat['internet'] :
    if x==0 :
        inter.append(0)
    elif x==1 :
        inter.append(1)
inter   
```

```python
dat['internet']=inter
```

```python
dat.head()
```

```python
gua=[]
for x in dat['guardian']:
    if x=="mother" :
        gua.append(2)
    elif x=="father" :
        gua.append(1)
    else:
        gua.append(0)
gua 
```

```python
dat['guardian']=gua
```

```python
dat.head()
```

```python
dat['caste'].value_counts()
```

```python
caste=[]
for x in dat['caste']:
    if x=="OC" :
        caste.append(1)
    elif x=="BC" :
        caste.append(2)
    elif x=="SC" :
        caste.append(3)
    elif x=="ST":
        caste.append(4)
caste
```

```python
dat['caste']=caste
```

```python
dat.head()
```

```python
dat.to_csv(r'modstudent.csv',header=True,index=None)
```
